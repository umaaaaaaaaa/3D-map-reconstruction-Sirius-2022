```python
import tensorflow_graphics.geometry.transformation as tfg
from hloc.utils import read_write_model
import numpy as np
```

```python
!mkdir -p /content/output_pose2
```

```python
# Transformation matrix 4x4(rotation + offset)
def build_matrix(x, y, z, R):
    M = np.zeros((4, 4))
    M[:3,:3] = R
    M[:, 3] = x, y, z, 1
    return M
```

```python
fname_template = "frame-{i:06}.pose.txt"
```

```python
imgs = read_write_model.read_images_binary("/images.bin")
```

```python
for i in range(1, 23):
  qvec = imgs[i].qvec
  tvec = imgs[i].tvec
  R = np.array(tfg.rotation_matrix_3d.from_quaternion(qvec))
  A = build_matrix(tvec[0], tvec[1], tvec[2], R)
  np.savetxt('/content/output_pose2/' + fname_template.format(i=i), A)
```
